# Milestone

Load csv file extract columns having json data and generate dataframe from json

```python
import pandas as pd
import json
```

# Load CSV

```python
df=pd.read_csv(r"C:\Users\Admin\Downloads\data.csv",header=None)
df.head()
```

# Add column name

```python
df.columns=["Value"]
df.head(100)
```

# Extract records with desired json

```python
index_list = [6,14,22]
new_df=df.loc[df.index[index_list]]
new_df.head()
```

# Reset index

```python
new_df.reset_index(drop=True, inplace=True)
new_df.head()
```

# Process json data

```python
dict_list=[]
for i in range(0,len(new_df)):
    
    json_str=dict(new_df.iloc[i])
    new_str=json_str['Value']
    new_dict=json.loads(new_str)
    dict_list.append(new_dict)
```

# Final Output

```python
final_df=pd.DataFrame(dict_list)
final_df.head()
```
